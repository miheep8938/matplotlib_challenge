## Observations and Insights 

This cancer study has investigated how effective each drug regimen would be for the group of mice that have cancerous tumors. Before analyzing the treatment's effectiveness, there was a mouse that contains two different data in terms of tumor volume and metastatic sites at the same timepoint. Unless I myself had done this experiment, there was no way to identify which values were actually examined in the experiment. This is the reason why I did not include the data related to the mouse 'g989'. 

In this data, the total number of mouse sample is 248 and the mice have been allocated into 10 different groups: 9 groups in different types of drugs, 1 group in control group. The mice sample is composed of 51% of male mice and 49% of female mice. 

According to the summary statistics and boxplot section, they show the general statistics the relationship between the tumor volumes and each drug. Noticeably, the groups of Capomulin and Ramicane that have the most number of mice have been effective in reducing tumor volumes in general. Compared to other treatments, Capomulin has the lowest standard deviation that indicates this treatment has reduced their tumor volumes for most of the mice in the group. In term of the mean volume of tumor, Ramicane has shown the lowest volumes. These data proves that these two drugs have been most effective in reducing tumor volumes. 

Another interesting fact that I have found is some of the drugs including Ketapril and Naftisol would have as effective as the placebo group/control group. In other words, these drugs could be considered as ineffective treatment for this tumor.  
In the box plot, there is the red dot in the Infubinol group representing that there is a single mouse’s tumor volume does not belong between upper bound and lower bound. The mouse, ‘c326’, has the volume of 36.32 that belongs to outlier since the lower bound of Infubinol is 36.83. 

In order to observe how much the tumor volume has reduced with Capomulin treatment, I selected one of the mice in the group, ‘s185’. Through the time, it shows gradual decline when it comes to the volume. 

Lastly, the scatter plot, correlation, and linear regression have shown the relationship between between the weight of each mouse in the Capomulin group and their tumor volume.  In the scatter plot, a tendency could be found that the heavier mice tend to have heavier tumor volumes. Plus the correlation coefficient value, 0.84, enhances its argument that there is a strong correlation between the weight of mice in the Capomulin group and their tumor size.  Moreover, as shown in the regression model, most of the variation in the response variable are around its mean. Overall, the data related to Capomulin and tumor size has shown the significant relationship. 


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset (day two, activity three)
merged_data = pd.merge(mouse_metadata,study_results, on="Mouse ID")
#merge.to_csv("output csv", index=False)

# Display the data table for preview
merged_data
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

#First, group the data by drug regimen

grouped_drug_regimen = merged_data.groupby('Drug Regimen')

#Find the mean value of Tumor Volume based on Drug Regimen 
mean_tumor_volume = grouped_drug_regimen['Tumor Volume (mm3)'].mean()

#Find the median value of Tumor Volume based on Drug Regimen 
median_tumor_volume = grouped_drug_regimen['Tumor Volume (mm3)'].median()

#Find the variance of Tumor Volume based on Drug Regimen 
variance_tumor_volume = grouped_drug_regimen['Tumor Volume (mm3)'].var()

#Find the Standard Deviation of Tumor Volume based on Drug Regimen 
sd_tumor_volume = grouped_drug_regimen['Tumor Volume (mm3)'].std()

#Find the Standard error of the mean of Tumor Volume 
sem_tumor_volume = grouped_drug_regimen['Tumor Volume (mm3)'].sem()

#Check those values above 
# print(mean_tumor_volume)
# print(median_tumor_volume)
# print(variance_tumor_volume)
# print(sd_tumor_volume)
# print(sem_tumor_volume)
```

```python
# Assemble the resulting series into a single summary dataframe.

summary_df = pd.DataFrame({
    "Mean": mean_tumor_volume, "Median": median_tumor_volume, "Variance": variance_tumor_volume,
    "Standard Deviation": sd_tumor_volume, "SEM": sem_tumor_volume
})
summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line

summary_agg= clean_data.groupby("Drug Regimen")
summary_2 = summary_agg.agg(['mean','median','var','std','sem'])["Tumor Volume (mm3)"]
summary_2
```

## Quartiles, Outliers and Boxplots

```python
#Additional: See the whole list 
list(summary_df.index)
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
regimen_df = clean_data[(clean_data["Drug Regimen"] == "Capomulin") |
                        (clean_data["Drug Regimen"] == "Ramicane")  |
                        (clean_data["Drug Regimen"] == "Infubinol") |
                        (clean_data["Drug Regimen"] == "Ceftamin")
                       ]

max_timepoint = regimen_df.groupby("Mouse ID")["Timepoint"].max()
max_timepoint_df = pd.DataFrame(max_timepoint)

#merge 
merge_drug = pd.merge(max_timepoint_df, clean_data, on=("Mouse ID","Timepoint"), how = 'left')
merge_drug_df = pd.DataFrame(merge_drug)
merge_drug_df

#Understand that some of the mice have no full timepoint. This represents that some of the old mice would be dead in the middle of the treatment.
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Create empty list to fill with tumor vol data (for plotting)
treatment_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
tumor_vol_data_list = []

for treatment in treatment_list: 
    tumor_vol_data = merge_drug.loc[merge_drug["Drug Regimen"] == treatment,'Tumor Volume (mm3)']   
    #print(tumor_vol_data)
   
    tumor_vol_data_list.append(tumor_vol_data)
    #print(tumor_vol_data_list)

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    quartiles = tumor_vol_data.quantile([0.25, 0.5, 0.75])
    first_quartile = quartiles[0.25]
    third_quartile = quartiles[0.75]
    iqr = third_quartile - first_quartile 
    # Determine outliers using upper and lower bounds
    lower_bound = first_quartile - (1.5*iqr)
    upper_bound = third_quartile + (1.5*iqr)
    print(f"The first quartile of {treatment} is {first_quartile}")
    print(f"The third quartile of {treatment} is {third_quartile}")
    print(f"The iqr value of {treatment} is {iqr}")
    print(f"The mean value of {treatment} is {quartiles[.5]}")
    print(f"The outliers of {treatment} would be below {lower_bound}")
    print(f"The outliers of {treatment} would be above {upper_bound}")
    print("---------------------------------------------------------- \n")
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig, ax = plt.subplots()
ax.set_title('Final Tumor Volume per Regimen')
ax.set_ylabel('Tumor Volume (mm3)')
props = dict(markerfacecolor = 'red', markersize = 12)
plt.boxplot(tumor_vol_data_list,labels=treatment_list, flierprops = props)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(avg_capo_vol['Weight (g)'],avg_capo_vol['Tumor Volume (mm3)'])
print(f"The correlation is {round(correlation[0],2)}")
```

```python
# Calculate linear regression model  
fig3,ax3 = plt.subplots()
x_values = avg_capo_vol['Weight (g)']
y_values = avg_capo_vol['Tumor Volume (mm3)']
(slope,intercept,rvalue,pvalue,stderr) = st.linregress (x_values,y_values)
regress_values = x_values *slope + intercept 
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Mouse Weight vs Average Tumor Volume')
plt.ylabel('Average Tumor Volume (mm3)')
ax3.annotate(line_eq, xy=(20, 40), xycoords='data',xytext=(0.8, 0.95), textcoords='axes fraction',horizontalalignment='right', verticalalignment='top',fontsize=15,color="red")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
